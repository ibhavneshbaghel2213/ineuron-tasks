```python
file = open("example.txt",'r+')
file.write("This is a placement assignment")
file.close()
```

```python
def placement_replacer(filename):
    try:
        with open(filename, "r+") as file:
            string = ""
            for i in file.read().split():
                if i == "placement":
                    i = "screening"
                string+=i+" "
            file.seek(0)
            file.write(string)
            file.close()
    except Exception as E:
        raise E
```

```python
placement_replacer("example.txt")
```

## 2. Demonstrate use of abstract class, multiple inheritance and decorator inpython using examples.

<b>Abstract class</b> : An abstract class is the class which contains one or more abstract methods. An abstract method is the one which is just defined but not implemented. every abstract class in Python should be derived from the ABC class which is present in the abc module and Object creation is not possible for abstract class only We can create objects for child classes of abstract classes to access implemented methods.

Example..,

```python
from abc import ABC, abstractmethod

class smartphone(ABC):
    
    def Smartphone_specification(self):   
        print("Ram, Rom, camera, etc..")
        
    @abstractmethod          
    def operating_system(self):    #abstract method
        pass
    
class Android(smartphone):   #child class of abstract class
    def operating_system(self):
        print("Most versions of operating system in android are available")
    
ob1 = Android()
ob1.operating_system()
```

<b>Multiple inheritance</b> : When you inherit a child class from more than one base classes, that situation is known as Multiple Inheritance. however, exhibits the same behavior as does the single inheritance beacause the child class claims the properties and methods of all the parent classes.

Example..,

```python
class Area():
    def get_Area(self, l, b):
        return l*b

class Perimeter():
    def get_Perimeter(self, l, b):
        return 2*(l + b)

class Rectangle(Area, Perimeter):
    def __init__(self, l, b):
        self.length = l
        self.breadth = b

    def area(self):
        return super().get_Area(self.length, self.breadth)

    def perimeter(self):
        return super().get_Perimeter(self.length, self.breadth)
    
Rect = Rectangle(10, 9)

print("Area:", Rect.area())
print("Perimeter:", Rect.perimeter())
```

<b>decorator</b> : Python decorators allow you to change the behavior of a function without modifying the function itself.In Decorators, functions are passed as an argument into another function and then called inside the wrapper function.

Example..,

```python
class Student:  
    def __init__(self,name,marks):  
         self.name = name  
         self.marks = marks 
        
    @property  
    def display(self):  
         return self.name + " got grade " + self.marks 
  
S = Student("Bhavnesh Baghel","A")  
print("Name:", S.name)  
print("grade:", S.marks)  
print(S.display)  
```
